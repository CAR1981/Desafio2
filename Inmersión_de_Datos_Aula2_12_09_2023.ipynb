Si tenemos el split () sin espesificar ningún parámtro, esto quiere decir que va a tomar automáticamente el espacio en blanco y en ese espacio en blanco va a hacer la separación. Por lo tanto va a devolver dos elementos en una lista. La función split() lo que hace es separar una cadena de texto en función de algún parámetro que le pasemos, por defecto toma el espacio

```python
inmuebles.Valor[0].split()
```

Ahora aplicamops la función para toda la columna

```python
inmuebles.Valor.str.split()
```

```python
type(inmuebles.Valor.str.split())
```

El campo uno (1) va a tener el valor númerico

```python
inmuebles.Valor.str.split(expand=True)
```

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

Como precio todavia es un string ,la función suma en lugar de hacer una suma entre los valores ha realizado una concatenación por tratarse de un string

```python
inmuebles['Precio'].sum()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio','Barrio']]
```

```python
inmuebles.info()
```

El siguiente paso es transformar el dato en float

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

%.2f = es el formato

lambda x = es una función que se aplica cuando es algo rápido, nos permite especificar algo muy sencillo, sirve para ajustar los valores.

```python
pd.set_option('display.precision',2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
inmuebles.describe()
```

loc es un método que nos permite localizar atributos que cumplan ciertas condiciones

```python
inmuebles.loc[inmuebles.Habitaciones == 110]
```

```python
columnas = {'Baños':'Banos','Área':'Area'}
inmuebles = inmuebles.rename(columns=columnas)
inmuebles.sample(10)
```

```python
inmuebles.loc[inmuebles.Area == 2]
```

```python
inmuebles['Precio_Millon'].plot.hist(bins=10, title='Precio_Millon', color='green')
```

matplotlib y seaborn son dos bibliotecas de python que en conjunto ayudar en la parte gráfica.

kde = significa curva de densidad


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de Valores de los inmuebles en Bogotá')
plt.xlim((50,1000))
#plt.ylim((0,20))
plt.savefig('/content/drive/MyDrive/Inmersión de datos/valor_inmuebles.png',format='png')
plt.show()
```

```python
tipos = ['Bodega', 'Lote','Finca']
inmuebles_tipos = inmuebles[inmuebles['Tipo'].isin(tipos)]

inmuebles_tipos
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
grafica = sns.histplot(data=inmuebles_tipos, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de Valores de los inmuebles en Bogotá, Colombia')
plt.xlim((30,6000))
#plt.ylim((0,20))
plt.savefig('/content/drive/MyDrive/Inmersión de datos/tipo_inmueble1.png',format='png')
plt.show()
```
